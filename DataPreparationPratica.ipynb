```python
# bibliotecas

import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv("/content/drive/MyDrive/PosGraduacao/CRISP-DM/data/DataPrepFinal.csv")
```

```python
df.head()
```

```python
df_clean = df.drop(columns = ['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
```

```python
df_clean.head()
```

```python
df_clean.dtypes
```

```python
df_clean['usd_goal_real'] = df_clean['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
#usando replace com regex
df_clean['launched'] = df_clean['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
df_clean.head()
```

```python
df_clean['usd_goal_real'] = df_clean['usd_goal_real'].astype('int')
```

```python
df_clean.dtypes
```

```python
df_clean['deadline']  = pd.to_datetime(df_clean['deadline'], format='%d/%m/%y')
df_clean['launched']  = pd.to_datetime(df_clean['launched'], format='%d/%m/%y')
df_clean.dtypes
```

```python
df_clean['time_range'] = (df_clean['deadline'] - df_clean['launched']).dt.days
df_clean
```
